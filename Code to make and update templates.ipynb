```python
def program_graph(program_choice, program_name, all_upper = False, render=True):

    elective_color = "#F2C695"
    requirement_color = "#ACD2A3"
    pre_req_color = "#E7A79B"
    dark_edge_color = "#030A10"
    non_eos_course = "#F5EDC9"
    edges = []
    
    one_of_color = "#E9DBD0"
    two_of_color = "#D1B79E"
    three_of_color = "#AD8152"
    
    one_of_color = "#dfe6e9"
    two_of_color = "#b2bec3"
    three_of_color = "#636e72"
    
    prenodecolors = [one_of_color,two_of_color,three_of_color]

    course_path = "data/courses/"
    program_path = "data/programs/"
    simplification_path = "data/simplifications/"
    
    course_files = os.listdir(course_path)
    course_files = list(set([c for c in course_files if "~" not in c]))

    program_df = pd.read_csv(program_path + f"{program_choice}")
    colors = sns.color_palette("deep",n_colors=len(list(program_df))).as_hex()
#     requirement_color = colors[0]
    
    program_courses = list(set(program_df.iloc[1:, :].values.ravel()))
    program_courses = [p for p in program_courses if type(p) is not type(2.341)]
    
    if any(['units' in d for d in program_df.values[0]]):
        available_color=non_eos_course
    else:
        available_color=non_eos_course

    u = graphviz.Digraph(
        "unix",
        filename="flow",
        format="svg",
        engine="dot",
        graph_attr={
#             "ranksep": ".6",
            "nodesep": "0.1",
#             "concentrate": "true",
            "splines": "spline",
            "label": f"{program_name}",
            "newrank":"true",
#             "landscale":"true",
            "labelloc":"t",
        },
        node_attr={
            "color": pre_req_color,
            "style": "filled",
            "fontsize": "9",
            "fontcolor": "black",
            "height": "0.1",
        },
    )

    i = 0
    
    with u.subgraph(
            name=f"cluster_Legend",
            graph_attr={"label": f"Legend"},
        ) as s:
        
        s.node('Required', shape="rectangle",
                    style="filled,rounded",
                    color=dark_edge_color,
                    fillcolor=requirement_color,
              width="2",)
        
        
        k=0
        for d in list(program_df)[1:]:
            N=program_df[d].dropna()[0]
            s.node(f'{d}{N} Required', shape="rectangle", label = f"{N} Required",
                    style="filled,rounded",
                    color=dark_edge_color,
                    fillcolor=colors[k],width="2")
            k+=1
            
        if all_upper:    
            s.node('Accessible EOS300+\nelectives', shape="rectangle",
                        style="filled,rounded",
                        color=dark_edge_color,
                        fillcolor=available_color,
                  width="2",)

#     for course in program_df[list(program_df)[0]].dropna()[1:]:
#         u.attr(
#             "node",
#             shape="rectangle",
#             style="filled,rounded",
#             color=dark_edge_color,
#             fillcolor=requirement_color,
#         )
#         u.node(course)

#     i = 0
#     for k in list(program_df)[1:]:
#         with u.subgraph(
#             name=f"cluster_{k}",
#             graph_attr={"label": f"Program requires: {program_df[k].dropna()[0]}"},
#         ) as s:
#             for course in program_df[k].dropna()[1:]:
#                 s.attr(
#                     "node",
#                     shape="rectangle",
#                     style="filled,rounded",
#                     color=dark_edge_color,
#                     fillcolor=colors[i],
#                 )
#                 s.node(course)
#         i += 1

    for course in course_files+program_courses:
        cs = course.split(".")[0]
        if cs in program_courses:
            column = np.where(program_df.isin([cs]).to_numpy())[1][0]
            if column == 0:
                cc = requirement_color
            else:
                cc = colors[column-1]
            u.node(
                cs,
                shape="rectangle",
                style="filled,rounded",
                color=dark_edge_color,
                fillcolor=cc,
                fontsize="9",
            )
        else:
            cc = "white"
        
#         all_courses.append(cs)

    if not all_upper:
        filtered_courses = program_courses
    else:
        filtered_courses = [e.split('.')[0] for e in course_files if (("EOS3" in e) or ("EOS4" in e))]
        options = set(program_courses)
        available_list = []
        for course in filtered_courses:
            u.attr(
                    "node",
                    shape="rectangle",
                    style="filled,rounded",
                    color=dark_edge_color,
                    fillcolor=available_color,
                    fontsize="9",
                    height="0.1",
                )
            df = pd.read_csv(course_path + course + ".csv")
            accessible = True
            for d in df:
                pre = set(df[d].dropna().values[1:])
                num = int(df[d].dropna().values[0])
                if len(options.intersection(pre)) < num:
                    u.attr(
                        "node",
                        shape="rectangle",
                        style="filled,rounded",
                        color=dark_edge_color + "20",
                        fillcolor="white",
                        fontsize="5",
                        height="0.01",
                    )
                    accessible = False

            if accessible:
                available_list.append(course)
                u.node(course)
        filtered_courses += program_courses

    u.attr(
        "node",
        shape="rectangle",
        style="filled,rounded",
        color=dark_edge_color + "20",
        fillcolor="white",
        fontsize="5",
        height="0.01",
        width="0.01",
    )

    course_files = os.listdir(course_path)
    course_files = [c for c in course_files if "~" not in c]
    courses = {}
    for c in course_files:
        course = c[:-4]
        courses[course] = {}
        df = pd.read_csv(course_path + course + ".csv")
        courses[course]["N Required"] = [df[k][0] for k in df.keys()]
        courses[course]["Options"] = [
            list(df[k][1:].dropna().values) for k in df.keys()
        ]


#     print(filtered_courses)
    
    for course in filtered_courses:
        if course in courses:
            for subset, N in zip(
                courses[course]["Options"], courses[course]["N Required"]
            ):
                if len(subset) == int(N):
                    for s in subset:
                        if s+course not in edges:
                            u.edge(
                            s,
                            course,
            #                 style=styles[color],
            #                 color=two_of_colors[color] + alpha_add,
                            penwidth="1",
                            arrowsize="0.5",
                            )
                            edges.append(s+course)
                else:
                    req = f"{N} of {subset}"
                    u.node(
                        req,
                        label=f"{N}",
                        fillcolor=prenodecolors[int(N) - 1],
                        fontsize="12",
                        fixedsize="true",
                        color=dark_edge_color,
                        shape="circle",width="0.3",height="0.3"
                    )
                    if req+course not in edges:
                        u.edge(
                        req,
                        course,
        #                 style=styles[color],
        #                 color=two_of_colors[color] + alpha_add,
                        penwidth="1",
                        arrowsize="0.5",
                        )
                    edges.append(req+course)
                    for s in subset:
                        if s+req not in edges:
                            u.edge(
                            s,
                            req,
            #                 style=styles[color],
            #                 color=two_of_colors[color] + alpha_add,
                            penwidth="1",
                            arrowsize="0.5",
                            )
                            edges.append(s+req)
        else:
            print(course)

                   
    u.attr(rankdir="LR")
    if render:
        if not all_upper:
            u.render(
                directory="source/programs/graphs/",
                filename=program_choice.split(".")[0],
                cleanup=True,
            )
        else:
            u.render(
                directory="source/programs/graphs/",
                filename=program_choice.split(".")[0]+'_upper',
                cleanup=True,
            )
    return u,colors,edges
```
